```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
```

```python
corpus = ['The series depicts the story of three poor sisters who are close',
          'their involvement in the case of 70 billion won which goes missing',
          'how they bravely face the richest family who are their opponents']
```

```python
bigram_vect = CountVectorizer(ngram_range=(2,2))
bigram_vect_matrix = bigram_vect.fit_transform(corpus)

bigram_vect_matrix = bigram_vect_matrix.toarray()
bigram_vocab = bigram_vect.get_feature_names()
pd.DataFrame(bigram_vect_matrix, columns=bigram_vocab)
```
